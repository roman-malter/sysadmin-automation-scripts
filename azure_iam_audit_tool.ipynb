```python
# Simulated cloud resources and their monthly costs
cloud_infrastructure = {
    "Web_Server_A": 45.00,
    "Database_Prod": 120.00,
    "Storage_Bucket": 15.50,
    "AI_Model_Test": 10.00
}

# The absolute maximum budget allowed by the company
MAX_BUDGET = 300.00
total_cost = 0.00

print("--- CLOUD BUDGET AUDIT RUNNING ---")

# Loop through resources and calculate total spending
for resource, cost in cloud_infrastructure.items():
    print(f"Checking {resource}: ${cost:.2f}")
    total_cost += cost

print("-" * 35)
print(f"TOTAL CURRENT SPEND: ${total_cost:.2f}")
print(f"BUDGET LIMIT: ${MAX_BUDGET:.2f}")
print("-" * 35)

# Evaluation Logic (Trial and Error check)
if total_cost > MAX_BUDGET:
    over_budget_amount = total_cost - MAX_BUDGET
    print(f"🚨 ALERT: Cloud environment is OVER BUDGET by ${over_budget_amount:.2f}!")
    print("Action Required: Shut down AI_Model_Test immediately to mitigate costs.")
else:
    print("✅ SUCCESS: Cloud environment is within safe budget limits.")
```
